#### Comparison of sequences

- Generate two DNA sequences (seq1 and seq2) with lenght of 500 letters using the routine you developed in the homework.

- Generate a third sequence based on the following rules.
    1. Use the same letter as seq1 and seq2 if two sequences are the same at the same position
    2. Use '-' if two sequences are different at the same position.

For example, if seq1 and seq2 are

`seq1: ACGT...TTA`

`seq2: ACTA...TAG`

Then seq3 should be

`seq3: AC--...T--`

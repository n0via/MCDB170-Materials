```python
def plot_skew(seq):
    GC_skew = [0]*len(seq)
    for idx in range(len(seq)):
        if seq[idx] == 'G':
            GC_skew[idx] = GC_skew[idx-1] + 1
        elif seq[idx] == 'C':
            GC_skew[idx] = GC_skew[idx-1] - 1
        else:
            GC_skew[idx] = GC_skew[idx-1]

    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()
    ax.plot(GC_skew)
```

## Vibrio Cholerae

```python
filename = 'GCF_000006745.1_ASM674v1_genomic.fna' # Examine this file yourself.

lines = open(filename,'r').readlines()
cholerae_chr1 = ''.join([  line.strip() for line in lines[1:37016]  ]) 
cholerae_chr2 = ''.join([  line.strip() for line in lines[37017:]  ]) 

plot_skew(cholerae_chr1)
```

```python
plot_skew(cholerae_chr2)
```
